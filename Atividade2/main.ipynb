```python
from tensorflow.keras import utils

path = './Data'
batch_size = 128

train = utils.image_dataset_from_directory(
    directory=path + '/train',
    shuffle = True,
    seed = 1,
    image_size = (32,32),
    batch_size = batch_size
)

test = utils.image_dataset_from_directory(
    directory=path + '/test',
    shuffle = True,
    seed = 1,
    image_size = (32,32),
    batch_size = batch_size
)
```

```python
from tensorflow.keras import models, layers, activations, initializers

model = models.Sequential([
    layers.Input(shape=(32,32,3)),

    layers.Conv2D(8, kernel_size=(3,3), strides=(1,1), padding='same'),
    # layers.MaxPooling2D((2,2)),

    layers.Conv2D(6, kernel_size=(3,3), strides=(1,1), padding='same'),
    layers.MaxPooling2D((2,2)),

    layers.Conv2DTranspose(12,kernel_size=(3,3), strides=(1,1), padding='same',activation=None,dilation_rate=(3, 3)),
    # layers.MaxPooling2D((2,2)),
    layers.Conv2DTranspose(10,kernel_size=(3,3), strides=(1,1), padding='same',activation=None,dilation_rate=(3, 3)),
    layers.MaxPooling2D((2,2)),


    layers.Flatten(), 

    layers.Dense(80, activation=activations.relu, kernel_initializer=initializers.RandomNormal()), 
    layers.Dropout(0.2),
    layers.Dense(60, activation=activations.relu, kernel_initializer=initializers.RandomNormal()),
    layers.Dense(70, activation=activations.relu, kernel_initializer=initializers.RandomNormal()),
    layers.Dropout(0.2),
    layers.Dense(50, activation=activations.relu, kernel_initializer=initializers.RandomNormal()),
    layers.Dense(10, activation=activations.softmax, kernel_initializer=initializers.RandomNormal())
])

from tensorflow.keras import optimizers, losses, metrics

lr = 0.001

model.compile(
    optimizer = optimizers.Adam(
        learning_rate = lr    
    ),
    loss = losses.SparseCategoricalCrossentropy(),
    metrics = [ metrics.sparse_categorical_accuracy ]
)

from tensorflow.keras import callbacks

patience = 6
epochs = 100
model_path = "./model.keras"

model.fit(
    train,
    validation_data = test,
    epochs = epochs,
    verbose = True,
    
    callbacks = [
        callbacks.EarlyStopping(
            monitor = 'val_loss',
            patience = patience,
            verbose = 1
        ),
        callbacks .ModelCheckpoint(
            filepath = model_path,
            save_weights_only = False,
            monitor = 'loss',
            mode = 'min',
            save_best_only = True
        )
    ]
)

model.fit(
    train,
    validation_data = test,
    epochs = epochs,
    verbose = True,
    
    callbacks = [
        callbacks.EarlyStopping(
            monitor = 'val_loss',
            patience = patience,
            verbose = 1
        ),
        callbacks .ModelCheckpoint(
            filepath = model_path,
            save_weights_only = False,
            monitor = 'loss',
            mode = 'min',
            save_best_only = True
        )
    ]
)
```
